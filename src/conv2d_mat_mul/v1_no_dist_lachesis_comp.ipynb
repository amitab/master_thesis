```python
import numpy as np
import math
```

```python
imgs = np.loadtxt("images_1.np").reshape(1,2,6,6)
kernel = np.loadtxt("kernel_3.np").reshape(3,2,3,3)
bias = np.loadtxt("bias_3.np").reshape(3)
```

```python
channels, X, Y = imgs[0].shape
```

```python
p = 0

def get_x():
    return X + p * 2

def get_y():
    return Y + p * 2

def get_win_val(w, i, j, s, ch, k, img):    
    h_slides = math.floor((get_y()-k)/s) + 1
    v_slides = math.floor((get_x()-k)/s) + 1
    
    r = math.floor(w / h_slides)
    c = w % h_slides
    x = (r * s) + i - p
    y = (c * s) + j - p
    
    if x < 0 or y < 0 or x >= X or y >= Y:
        return 0
    
    return imgs[img][ch][x][y]

def img_to_conv2d():
    s = 1
    k = 3

    h_slides = math.floor((get_y()-k)/s) + 1
    v_slides = math.floor((get_x()-k)/s) + 1
    
    windows = h_slides * v_slides
    
    data = []
    
    for w in range(windows):
        row = []
        for c in range(channels):
            for i in range(k):
                for j in range(k):
                    row.append(get_win_val(w, i, j, s, c, k, 0))
        data.append(row)
    
    return np.array(data)
```

```python
data = img_to_conv2d()
data = np.hstack((data, np.ones(data.shape[0]).reshape(-1, 1)))
for i in range(data.shape[0]):
    print(data[i].tolist())
```

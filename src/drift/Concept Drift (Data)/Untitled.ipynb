```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import math
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(-2*np.pi, 2*np.pi, 30000)

arr = np.exp2(x)

start = 0
end = 1
width = end - start
res = (arr - arr.min())/(arr.max() - arr.min()) * width + start
res[res>0.5] = 0.5

act_space = np.linspace(0, 10, 60000)
act_res = np.ones(60000)
act_res[:30000] = res
act_res[30000:] = 0.5

ax.plot(act_space, act_res);
```
